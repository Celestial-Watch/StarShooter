```python
# WRite a function to get the image
def get_image(url, id_):
    r = requests.get(url, headers={"Cookie": "PHPSESSID=41d61821d1c671383cca68df56573419"})
    soup = BeautifulSoup(r.content, "html.parser")
    img = soup.findAll("img")[-1]
    img_url = base + img["src"]
    os.system(f"wget {img_url} -O images/{id_}.png")
```
