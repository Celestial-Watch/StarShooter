```python
import pandas as pd
```

```python
test_data_lookup = pd.read_csv("test_data_set.csv")
test_metadata = pd.read_csv("test_movers_meta_data.csv")
# make last column of metadata the first column
metadata_columns = test_metadata.columns.tolist()
metadata_columns = metadata_columns[-1:] + metadata_columns[:-1]
test_metadata = test_metadata[metadata_columns]
test_metadata.rename(columns={"Name": "mover_id"}, inplace=True)
```

```python
test_data_lookup_matching = test_data_lookup.copy()
test_data_lookup_matching["file_index"] = (
    test_data_lookup_matching["image"]
    .apply(lambda x: x.split("-")[-1].replace(".png", ""))
    .astype(int)
)
test_data_lookup_matching
```

```python
regex_match = {
    "A": 1,
    "B": 2,
    "C": 3,
    "D": 4,
}


def find_character_in_fit_file(fit_file: str):
    for key, value in regex_match.items():
        if key in fit_file:
            return value
    return None
```

```python
test_metadata["FileName"] = test_metadata["FileName"].apply(find_character_in_fit_file)
test_metadata.rename(columns={"FileName": "file_index"}, inplace=True)
test_metadata
```

```python
testset_with_metadata = test_data_lookup_matching.merge(
    test_metadata, on=["mover_id", "file_index"]
)
testset_with_metadata
```

```python
testset_with_metadata.drop(columns=["Link", "file_index"], inplace=True)
```

```python
testset_with_metadata.rename(
    columns={
        "image": "file_name",
    },
    inplace=True,
)
```

```python
testset_with_metadata.to_csv("test_data_set_with_metadata.csv", index=False)
```
