```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("models_data_processed.csv")
data
```

```python
same_learning_rate = data[data["learning_rate"] == 0.001]
same_learning_rate
```

```python
#Group by number_convs and kernel_size taking the mean and best of the values
grouped = same_learning_rate.groupby(["number_convs", "kernel_size"]).agg({"val_acc_best": ["mean", "max", "min"]})
#remove kernel size of 2
grouped = grouped.drop(2, level="kernel_size")

#Multiple val_acc_best by 100
grouped["val_acc_best"] *= 100
grouped
```

```python
# plot heat map of val_acc_best mean
grouped_mean = grouped["val_acc_best"]["mean"].unstack()
plt.figure(figsize=(10, 7))
sns.heatmap(grouped_mean, annot=True, fmt=".2f")
plt.title("Mean of val_acc_best with different number of convolutions and kernel size")
plt.xlabel("Kernel size")
plt.ylabel("Number of convolutions")
plt.show()
```

```python
# plot heat map of val_acc_best mean
grouped_mean = grouped["val_acc_best"]["max"].unstack()
plt.figure(figsize=(10, 7))
sns.heatmap(grouped_mean, annot=True, fmt=".2f")
plt.title("Peak of val_acc_best with different number of convolutions and kernel size")
plt.xlabel("Kernel size")
plt.ylabel("Number of convolutions")
plt.show()
```

```python
# plot heat map of val_acc_best mean
grouped_mean = grouped["val_acc_best"]["min"].unstack()
plt.figure(figsize=(10, 7))
sns.heatmap(grouped_mean, annot=True, fmt=".2f")
plt.title("Worst of val_acc_best with different number of convolutions and kernel size")
plt.xlabel("Kernel size")
plt.ylabel("Number of convolutions")
plt.show()
```

```python
# Group by feature vector size and take the mean and best of the values
feature_vector_size = same_learning_rate.groupby(["feature_vector_size"]).agg({"val_acc_best": ["mean", "max"]})
feature_vector_size
```

```python
#plot a barchat of mean and max beside each other for each feature vector size
feature_vector_size.plot(kind="bar", figsize=(10, 5))
plt.title("Mean and Max Validation Accuracy for Different Feature Vector Sizes")
plt.xlabel("Feature Vector Size")
plt.ylabel("Validation Accuracy")
plt.legend(["Mean", "Max"])
plt.ylim(0.92, 1)
plt.show()

```

```python
#Group by fully connected layers and units_per_layer taking the mean and best of the values
fully_connected_layers = same_learning_rate.groupby(["fully_connected_layers", "units_per_layer"]).agg({"val_acc_best": ["mean", "max"]})
fully_connected_layers
#remove units_per_layer with 256
fully_connected_layers = fully_connected_layers.drop(256, level="units_per_layer")
fully_connected_layers = fully_connected_layers * 100
fully_connected_layers
```

```python
fully_connected_layers["val_acc_best"]["max"].unstack()
```

```python
#use seaborn to plot a heatmap of the mean validation accuracy for different fully connected layers and units per layer to 4 significant figures
plt.figure(figsize=(10, 5))
sns.heatmap(fully_connected_layers["val_acc_best"]["max"].unstack(), annot=True, fmt=".2f")
plt.title("Max Validation Accuracy for Different Fully Connected Layers and Units per Layer")
plt.xlabel("Units per Layer")
plt.ylabel("Fully Connected Layers")
plt.show()
```

```python
# Group by channels_per_conv and feature vector and take the mean and best of the values
channels_per_conv = same_learning_rate.groupby(["channels_per_conv", "feature_vector_size"]).agg({"val_acc_best": ["mean", "max", "min"]})
channels_per_conv
```

```python
# get all the values for channels_per_conv and feature_vector_size with a 2 conv layers and a kernel size of 3, with a 2 fully connected layers and 64 units per layer
channels_feature = same_learning_rate[(same_learning_rate["number_convs"] == 1) & (same_learning_rate["kernel_size"] == 3) & (same_learning_rate["fully_connected_layers"] == 2) & (same_learning_rate["units_per_layer"] == 64)]
channels_feature["val_acc_best"] *= 100
channels_feature["val_acc_mean"] = channels_feature["val_acc_mean"] * 100
channels_feature

```

```python
#drop all the columns that are not needed
channels_feature = channels_feature.drop(["model_name", "number_convs", "kernel_size", "fully_connected_layers", "units_per_layer", "learning_rate"], axis=1)
channels_feature = channels_feature.drop(["train_loss_mean", "train_loss_best", "val_loss_mean", "val_loss_best"], axis=1)
channels_feature = channels_feature.drop(["val_acc_mean"], axis=1)
#sort by feature_vector_size
channels_feature = channels_feature.sort_values(["feature_vector_size", "channels_per_conv"])
channels_feature
```

```python
# reshape the dataframe to have feature_vector_size as the index and channels_per_conv as the columns
channels_feature = channels_feature.pivot(index="feature_vector_size", columns="channels_per_conv", values="val_acc_best")
channels_feature
```

```python
# plot heatmap of the max validation accuracy for different channels per conv and feature vector size
plt.figure(figsize=(10, 5))
sns.heatmap(channels_feature, annot=True, fmt=".2f")
plt.title("Max Validation Accuracy for Different Channels per Conv and Feature Vector Size for 1 Conv Layer, 3 Kernel Size, 2 Fully Connected Layers, 64 Units per Layer")
plt.xlabel("Channels per Conv")
plt.ylabel("Feature Vector Size")
plt.show()
```
