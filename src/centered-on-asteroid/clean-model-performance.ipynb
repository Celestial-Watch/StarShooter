```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
raw_data = pd.read_csv("models_data.csv")
raw_data.columns = ["model_name", "train_loss", "val_loss", "val_acc"]
# remove [] from train_loss and val_loss and val_acc
raw_data["train_loss"] = raw_data["train_loss"].str.strip("[").str.strip("]")
raw_data["val_loss"] = raw_data["val_loss"].str.strip("[").str.strip("]")
raw_data["val_acc"] = raw_data["val_acc"].str.strip("[").str.strip("]")
raw_data
```

```python
data = raw_data.copy()
data["train_loss_mean"] = data["train_loss"].apply(
    lambda x: np.mean([float(i) for i in x.split(",")])
)
data["train_loss_best"] = data["train_loss"].apply(
    lambda x: np.min([float(i) for i in x.split(",")])
)
data["val_loss_mean"] = data["val_loss"].apply(
    lambda x: np.mean([float(i) for i in x.split(",")])
)
data["val_loss_best"] = data["val_loss"].apply(
    lambda x: np.min([float(i) for i in x.split(",")])
)
data["val_acc_mean"] = data["val_acc"].apply(
    lambda x: np.mean([float(i) for i in x.split(",")])
)
data["val_acc_best"] = data["val_acc"].apply(
    lambda x: np.max([float(i) for i in x.split(",")])
)
data = data.drop(columns=["train_loss", "val_loss", "val_acc"])
data
```

```python
data.to_csv("models_data_processed.csv", index=False)
```

```python
data = pd.read_csv("models_data_processed.csv")
data
```

```python
model_names = data["model_name"]
model_names
```

```python
model_parameters = {}
model_parameter_list = [
    "number_convs",
    "kernel_size",
    "channels_per_conv",
    "feature_vector_size",
    "learning_rate",
    "fully_connected_layers",
    "units_per_layer",
]
for name in model_names:
    values_with_annotation = name.split("-")
    values = [i.split("_")[-1] for i in values_with_annotation]
    model_parameters[name] = dict(zip(model_parameter_list, values))

model_parameters
```

```python
# make model_parameters a dataframe
model_parameters_df = pd.DataFrame(model_parameters).T
model_parameters_df
```

```python
merged_data = pd.merge(
    model_parameters_df, data, left_index=True, right_on="model_name"
)
merged_data
```

```python
# move model_name to the first column
cols = merged_data.columns.tolist()
cols.insert(0, cols.pop(cols.index("model_name")))
merged_data = merged_data.reindex(columns=cols)
merged_data
```

```python
# make number_convs, kernel_size, channels_per_conv, fully_connected_layers, units_per_layer integers
merged_data["number_convs"] = merged_data["number_convs"].astype(int)
merged_data["kernel_size"] = merged_data["kernel_size"].astype(int)
merged_data["channels_per_conv"] = merged_data["channels_per_conv"].astype(int)
merged_data["fully_connected_layers"] = merged_data["fully_connected_layers"].astype(
    int
)
merged_data["units_per_layer"] = merged_data["units_per_layer"].astype(int)
merged_data["feature_vector_size"] = merged_data["feature_vector_size"].astype(int)
merged_data
```

```python
merged_data.to_csv("models_data_processed.csv", index=False)
```
