```python
import pandas as pd
```

```python
raw_movers = pd.read_csv("movers_images_lookup.csv")
print(f"The length of the raw movers data is {len(raw_movers)}")
raw_movers.head()
```

```python
movers_index = raw_movers["mover_id"].value_counts() == 4
movers_index = movers_index[movers_index].index
movers = raw_movers[raw_movers["mover_id"].isin(movers_index)]
print(f"The length of the cleaned movers data is {len(movers)}")
movers.head()
```

```python
print(f"The number of movers is {len(movers) / 4}")
```

```python
movers_for_matching = movers.copy()
movers_for_matching["file_index"] = movers_for_matching["file_name"].apply(
    lambda x: x.split("-")[1].replace(".png", "")
)
movers_for_matching["file_index"] = movers_for_matching["file_index"].astype(int)
movers_for_matching

movers_for_matching["label"] = 1
movers_for_matching
```

```python
unique_mover_ids = movers_for_matching["mover_id"].unique()
```

```python
# Rule for matching fit file to small image
regex_matxh = {
    "A": 1,
    "B": 2,
    "C": 3,
    "D": 4,
}


def find_character_in_fit_file(fit_file: str):
    for key, value in regex_matxh.items():
        if key in fit_file:
            return value
    return None


def match_fit_file_to_small_image(movers_table, mover_id, file_index):
    # Find if A, B, C, or D is in the fit file
    # Find the small image that matches the mover_id and the character is in the split on "-"
    filtered_movers = movers_table[movers_table["mover_id"] == mover_id]
    print(filtered_movers)
    # Return the file name with the file index that matches the fit file
    small_image = filtered_movers[filtered_movers["file_index"] == int(file_index)]
    return small_image
```

```python
regex_matxh.items()
```

```python
# import the mover metadata
raw_movers_metadata = pd.read_csv("movers_meta_data.csv")


# make last column the first column
cols = list(raw_movers_metadata.columns)
cols = [cols[-1]] + cols[:-1]
raw_movers_metadata = raw_movers_metadata[cols]

# Rename Name to mover_id
raw_movers_metadata = raw_movers_metadata.rename(
    columns={"Name": "mover_id", "FileName": "fit_file"}
)

# Filter to only have the unique movers
raw_movers_metadata = raw_movers_metadata[
    raw_movers_metadata["mover_id"].isin(unique_mover_ids)
]
print(f"The length of the raw movers metadata is {len(raw_movers_metadata)}")
raw_movers_metadata.head(20)
```

```python
raw_movers_metadata["fit_file"] = raw_movers_metadata["fit_file"].apply(
    lambda x: find_character_in_fit_file(x)
)
raw_movers_metadata
```

```python
raw_movers_metadata.rename(columns={"fit_file": "file_index"}, inplace=True)
```

```python
# Merge the movers metadata with the movers images using mover_id and file_index
movers_with_metadata = movers_for_matching.merge(
    raw_movers_metadata, on=["mover_id", "file_index"]
)
```

```python
movers_with_metadata
```

```python
whole_image_look_up = movers_with_metadata[["mover_id", "Link"]]
whole_image_look_up["Link"] = whole_image_look_up["Link"].apply(
    lambda x: x.split("=")[-1]
)
whole_image_look_up.rename(columns={"Link": "totas_id"}, inplace=True)
```

```python
whole_image_look_up
```

```python
whole_image_look_up.to_csv("final_csv/movers_whole_images_lookup.csv", index=False)
```

```python
# Drop file_index and link
movers_with_metadata = movers_with_metadata.drop(columns=["label"])
movers_with_metadata
```
